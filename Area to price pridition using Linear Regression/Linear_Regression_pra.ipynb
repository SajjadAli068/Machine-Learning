```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('area_price.csv')
df.head()
```

```python
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area Price")
plt.scatter(x=df['area'], y=df['price'] , color='red' , marker="*")
```

```python
# create the model
model = LinearRegression()
model.fit(df[['area']],df['price'])
```

```python
# to predict the value
model.predict([[5000]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
import seaborn as sns

df = sns.load_dataset('titanic')
df.head()
```

```python
df.isnull().sum()
```

```python
df_linear = df[['age','fare']]
df_linear.isnull().sum()
```

```python
df_linear['age'].fillna(df_linear['age'].mean() , inplace=True)
```

```python
df_linear.isnull().sum()
```

```python
# now create the model
model = LinearRegression()
model.fit(df_linear[['age']],df_linear['fare'])
```

```python
model.predict([[20]])
```

```python
model.coef_
```

```python
model.intercept_
```
