```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('area_price.csv')
df.head()
```

```python
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area Price")
plt.scatter(x=df['area'], y=df['price'] , color='red' , marker="*")
```

```python
# create the model
model = LinearRegression()
model.fit(df[['area']],df['price'])
```

```python
# to predict the value
model.predict([[5000]])
```

```python
model.coef_
```

```python
model.intercept_
```
