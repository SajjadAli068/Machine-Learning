```python
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
# I only take two column which name age and fare
df_linear = df[['age','fare']]
df_linear.isnull().sum()
```

```python
df_linear['age'].fillna(df_linear['age'].mean() , inplace=True)
```

```python
df_linear.isnull().sum()

sns.lmplot(df, x='age',y='fare')
```

```python
# create the model

model = LinearRegression()

# fitting

model.fit(df_linear[['age']],df_linear['fare'])

# predict 
model.predict([[80]])
```
