```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('book_1.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
# Create the model
model = linear_model.LinearRegression()
model.fit(df[['year']], df.income)
```

```python
# Predict 

model.predict([[2016]])
```
